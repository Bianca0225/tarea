```python
# Importamos las librer√≠as 
import yahoo_finance as yahoo
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from pandas_datareader import data
import seaborn as sns
import yfinance as yf
```

```python
#Realizaremos un dataframe que contenga todas las empresas de la industria de telecomunicaciones
empresas_sect=pd.read_excel("Empresas de la industria Telecomunicaciones.xlsx")
```

```python
tickers = empresas_sect["Symbol"].tolist()
```

```python
tickers
```

```python
dnuevo=pd.DataFrame()
for nuev in tickers:
    dnuevo[nuev]=yf.Ticker(nuev).history(start="2020-01-01",end="2023-8-25").Close
```

```python
#Realizaremos una gr√°fica que contenga los precios historicos de las empresas de la industria para un periodo peque√±o ya que son muchos datos
dnuevo.plot(figsize=(15,6))
plt.title('Historico de precios de las acciones del sector de Telecomunicaciones')
plt.ylabel('Precio en USD')
plt.xlabel('Fecha')
plt.show()
```

```python
# Ahora seleccionaremos las 25 acciones con mayor valor intrad√≠a del sector de telecomunicaciones
list_of_tickers = ["CABO","CHTR","NFLX","META","MSGS","NXST","TTWO","SPOT","BIDU","TMUS","GOOG","GOOGL","EA","WWE","NTES","ATVI","LBRDK","LYV","IDCC","DIS","OMC","DASH","ROKU","CCOI"]
```

```python
df=pd.DataFrame()
for ticker in list_of_tickers:
    df[ticker]=yf.Ticker(ticker).history(start="2020-1-1",end="2023-8-25").Volume
```

```python
df.plot(figsize=(12,6))
plt.title('Volumen de las 25 compa√±√≠as con mayor precio del sector Telecomunicaciones')
plt.ylabel('Volumen')
plt.xlabel('Fecha')
plt.show()
```

```python
#Ahora se ver√° el historico de precios de las 25 compa√±√≠as seleccionadas
df=pd.DataFrame()
for ticker in list_of_tickers:
    df[ticker]=yf.Ticker(ticker).history(start="2020-1-1",end="2023-8-25").Close

df.plot(figsize=(12,6))
plt.title('Historico de precios de las 25 compa√±√≠as con mayor precio del sector Telecomunicaciones')
plt.ylabel('Precio en USD')
plt.xlabel('Fecha')
plt.show()
```

```python
#Para una mejor visualizaci√≥n pondremos m√°s atenci√≥n en las empresas m√°s conocidas
lista_resumida= ["NFLX","META","SPOT","GOOG","GOOGL","DIS","ROKU"]
dfresumido=pd.DataFrame()
for tick in lista_resumida:
    dfresumido[tick]=yf.Ticker(tick).history(start="2020-1-1",end="2023-8-25").Volume
```

```python
dfresumido.plot(figsize=(12,6))
plt.title('Volumen de 7 de las compa√±√≠as m√°s conocidas en el sector Telecomunicaciones')
plt.ylabel('Volumen')
plt.xlabel('Fecha')
plt.show()
```

```python
#Ahora se ver√° el historico de precios de las 7 compa√±√≠as consideradas m√°s conocidas
for tick in lista_resumida:
    dfresumido[tick]=yf.Ticker(tick).history(start="2020-1-1",end="2023-8-25").Close
    
dfresumido.plot(figsize=(12,6))
plt.title('Historico de precios de las 7 de las compa√±√≠as m√°s conocidas en el sector Telecomunicaciones')
plt.ylabel('Precio en USD')
plt.xlabel('Fecha')
plt.show()
```

```python
# Ahora realizaremos un an√°lisis 3 de las empresas del sector que se pueden considerar competencia
lista_comp= ["NFLX","DIS","ROKU"]
dfcomp=pd.DataFrame()
for tickcomp in lista_comp:
    dfcomp[tickcomp]=yf.Ticker(tickcomp).history(start="2020-1-1",end="2023-8-25").Close
```

```python
dfcomp.plot(figsize=(12,6))
plt.title('Historico de precios de 3 acciones del sector que compiten entre si')
plt.ylabel('Precio en USD')
plt.xlabel('Fecha')
plt.show()
```

```python
# Ahora realizaremos un an√°lisis de las 5 empresas del sector con mayor Market Cap
lista_marketcap= ["GOOGL","GOOG","META","CMCSA","NFLX"]
dfmc=pd.DataFrame()
for MC in lista_marketcap:
    dfmc[MC]=yf.Ticker(MC).history(start="2020-1-1",end="2023-8-25").Close
```

```python
dfmc.plot(figsize=(12,6))
plt.title('Historico de precios de las 5 acciones con mayor Market Cap del sector')
plt.ylabel('Precio en USD')
plt.xlabel('Fecha')
plt.show()
```

```python
# Todos los analisis se han realizado en tiempo de pandemia y postpandemia, ahora consideraremos tambi√©n el periodo prepandemia de las 5 empresas del sector con mayor Market Cap
lista_marketcap= ["GOOGL","GOOG","META","CMCSA","NFLX"]
dfmc=pd.DataFrame()
for MC in lista_marketcap:
    dfmc[MC]=yf.Ticker(MC).history(start="2000-1-1",end="2023-8-25").Close
    
dfmc.plot(figsize=(12,6))
plt.title('Historico de precios de las 5 acciones con mayor Market Cap del sector')
plt.ylabel('Precio en USD')
plt.xlabel('Fecha')
plt.show()
```
